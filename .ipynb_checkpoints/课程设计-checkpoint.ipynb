```python
'''
@fn 函数名：isPrime
@brief 函数功能：素数判断算法
@data 函数输入：num 一个数字
@return 函数输出 是否是素数
'''
def isPrime(num):
    # 质数大于 1
    if num > 1:
       # 查看因子
       for i in range(2,num):
           if (num % i) == 0:
                return False
       else:
            return True
    # 如果输入的数字小于或等于 1，不是质数
    else:
       return False
'''
@fn 函数名：isCoPrime
@brief 函数功能：判断两个输入的数是否为互为素数
@data 函数输入：a 一个数字  b 第二个数字
@return 函数输出 是否是互素数
'''
def isCoPrime(a,b):
    while(b!=0):
        temp=b
        b=a%b
        a=temp
    if(a==1):
        return True
    else:
        return False
import sys
sys.setrecursionlimit(100000)   #设置迭代次数限制
'''
@fn 函数名：createKey
@brief 函数功能：输入两个素数，生成公钥和私钥
@data 函数输入：a 一个数字  b 第二个数字
@return 函数输出 公钥e 私钥d
'''
import random
def createKey(a,b):
    if(isPrime(a) and isPrime(b)):
        k=(a-1)*(b-1)
        n=a*b
        e=random.randint(2,k-1)#PublicKey
        while(not isCoPrime(e,k)):
            e=e+1
        d=1
        while((e*d)%k!=1):#这里有可能出现ed一开始就大于n的情况，所以余数永远是n
            d=d+1
        return (n,e),(n,d)
    else:
        print("plz give 2 Prime number")
        return 
'''
@fn 函数名：encrypt
@brief 函数功能：
加密算法
对输入的明文首先分块，然后利用公钥进行加密，生成密文。
m^e mod n ≡ c  
m必须是整数（字符串可以取ascii值或unicode值），且m必须小于n。
@data 函数输入：明文m(字符串)   公钥（e，n）
@return 函数输出 密文c
'''
def encrypt(m,PublicKey):
    e=PublicKey[1]
    n=PublicKey[0]
    lens=len(str(n))
    string=""
    for i in m:
        mi=ord(i)
        mi=mi**e%n
        stri=str(mi)
        if len(stri)<lens:
            for i in range(lens-len(stri)):
                string=string+'0'
        string=string+stri
    return string
'''
@fn 函数名：decrypt
@brief 函数功能：
解密算法
将步骤2算法中获得的密文利用私钥进行解密。
c^d mod n ≡ m  
@data 函数输入：密文c 私钥（d，n）
@return 函数输出 明文m
'''
def decrypt(c,PrivateKey):
    d=PrivateKey[1]
    n=PrivateKey[0]
    lens=len(str(n))
    string=""
    for i in range(0,len(c),lens):
        stri=int(c[i:i+lens])
        mi=(stri**d)%n
        string=string+chr(mi)
    return string
```

```python
PublicKey,PrivateKey=createKey(13,17)
m="Hello World"
c=encrypt(m,PublicKey)
print("c:",c)
m=decrypt(c,PrivateKey)
print("m:",m)
```
